```python
import pandas as pd
```

```python
url = 'https://github.com/HamoyeHQ/HDSC-Introduction-to-Python-for-machine-learning/files/7768140/FoodBalanceSheets_E_Africa_NOFLAG.csv'
fbs = pd.read_csv(url, encoding = "latin-1")
```

```python
fbs.describe(include = 'all')
```

```python
fbs.groupby(['Y2017', 'Area'])['Area'].sum()
```

```python
fbs.groupby('Item')['Item'].sum()
```

```python
fbs.groupby(['Y2015', 'Item'])['Item'].agg(sum)
```

```python
fbs.groupby('Area')['Area'].count()
```

```python
fbs.groupby(['Y2017', 'Area'])['Area'].agg([min, max, sum])
```
